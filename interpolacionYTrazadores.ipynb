```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange, CubicSpline

# Datos de la tabla
puntos_x = np.array([0, 1, 2, 3, 4, 5, 6])
puntos_y = np.array([3, 5, 6, 5, -2, 2, 13])

# Punto de evaluación
x_eval = 3.25

# ---- Polinomio de Lagrange ----
polinomio_lagrange = lagrange(puntos_x, puntos_y)
y_lagrange = polinomio_lagrange(x_eval)

# ---- Trazador cúbico (Spline cúbico) ----
spline_cubico = CubicSpline(puntos_x, puntos_y)
y_spline = spline_cubico(x_eval)

# ---- Gráficas ----
x_vals = np.linspace(0, 6, 100)
y_vals_lagrange = polinomio_lagrange(x_vals)
y_vals_spline = spline_cubico(x_vals)

plt.figure(figsize=(10, 6))

# Gráfica del polinomio de Lagrange
plt.plot(x_vals, y_vals_lagrange, label="Polinomio de Lagrange", color="blue")
# Gráfica del Spline cúbico
plt.plot(x_vals, y_vals_spline, label="Trazador cúbico", color="green")
# Puntos dados
plt.scatter(puntos_x, puntos_y, color="red", label="Puntos dados")

# Mostrar puntos de evaluación
plt.scatter(x_eval, y_lagrange, color="blue", marker="o", label=f"f_Lagrange(3.25) ≈ {y_lagrange:.2f}")
plt.scatter(x_eval, y_spline, color="green", marker="x", label=f"f_Spline(3.25) ≈ {y_spline:.2f}")

# Configuración de la gráfica
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Interpolación usando Polinomio de Lagrange y Trazadores Cúbicos")
plt.legend()
plt.grid(True)
plt.show()

# Resultados
print(f"Valor estimado de f(3.25) usando Polinomio de Lagrange: {y_lagrange:.4f}")
print(f"Valor estimado de f(3.25) usando Trazador cúbico: {y_spline:.4f}")
```
